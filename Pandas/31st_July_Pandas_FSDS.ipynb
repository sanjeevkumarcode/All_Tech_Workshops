```python
def profit_flag(a):
    if a>0:
        return 'positive'
    else:
        return 'negative'
```

```python
df_sales['flag_profit']= df_sales['Profit'].apply(profit_flag)
```

```python
def evenodd (b):
    if(b%2==0):
        return 'even'
    else:
        return 'odd'
```

```python
evenodd(12)
```

```python
evenodd(13)
```

```python
df_sales['oddeven']=df_sales['Customer ID'].apply(evenodd)
```
